```python
import numpy as np

# Create array
arr = np.array([10, 20, 30, 40, 50])
print("Original array:", arr)

# Reshape
reshaped = arr.reshape((5, 1))
print("Reshaped array:\n", reshaped)

# Basic statistics
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
print("Max:", np.max(arr))
print("Min:", np.min(arr))
```

```python
import pandas as pd

data = {
    "student": ["Asha", "Ben", "Chen", "Dina", "Ethan"],
    "mark":    [78, 92, 65, 84, 55]
}

df = pd.DataFrame(data)
df
```

```python
# Save to CSV (in the same folder as the notebook)
df.to_csv("sample_marks.csv", index=False)

# Load from CSV
df2 = pd.read_csv("sample_marks.csv")
df2
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(df2["student"], df2["mark"], marker="o")
plt.title("Student Marks")
plt.xlabel("Student")
plt.ylabel("Mark")
plt.ylim(0, 100)
plt.grid(True)
plt.show()
```
